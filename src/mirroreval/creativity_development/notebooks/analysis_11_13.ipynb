```python
import pandas as pd
import json
```

```python
def load_jsonl(path):
    """Load a JSONL file into a pandas DataFrame."""
    with open(path, "r", encoding="utf-8") as f:
        data = [json.loads(line) for line in f]
    return pd.DataFrame(data)
```

```python
df = load_jsonl("/home/jayo/repos/MIRROR-Eval/results/testing/results.jsonl")
```
